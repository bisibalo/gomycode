```python
from bs4 import BeautifulSoup
```

```python
soups = BeautifulSoup(blinks.content, 'html.parser')
```

```python
soups.prettify()
```

```python
# Write a function to Extract article title
soups.title.text
```

```python
#  Write a code to Extract article text for each paragraph with their respective
# headings. Map those headings to their respective paragraphs in the dictionary.

soups.find_all("th")
```

```python
header =[]
for i in soups.find_all("th"):
    header.append(i.text)
```
