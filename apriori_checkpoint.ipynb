```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
```

```python
te = TransactionEncoder()
```

```python
te_array = te.fit_transform(toy_dataset)
te_array
```

```python
df = pd.DataFrame(te_array, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori as ap
from mlxtend.frequent_patterns import apriori
```

```python
items = set(item for transaction in toy_dataset for item in transaction)

df = pd.DataFrame([{item: (item in transaction) for item in items} for transaction in toy_dataset])

print("One-Hot Encoded Dataset:\n", df)
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print("\nFrequent Itemsets:\n", frequent_itemsets)
print("\nAssociation Rules:\n", rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
