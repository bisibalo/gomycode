```python
numeric_cols = ['ZIP Code', 'Model Year', 'Electric Range', 'Base MSRP', 'Legislative District', 'DOL Vehicle ID']

Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df[~((df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound)).any(axis=1)]
```

```python
category = df_clean[['ID', 'VIN (1-10)','County', 'City', 'State','Make','Model','Electric Vehicle Type','Clean Alternative Fuel Vehicle (CAFV) Eligibility'
               ,'Vehicle Location', 'Electric Utility']]
```

```python
dummies = pd.get_dummies(df_clean.State, drop_first=True)
```

```python
dummies
```

```python
merged = pd.concat([df_clean, dummies], axis=1)
```

```python
df = merged.drop(columns=['City','Make','Model','Electric Vehicle Type', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility','Vehicle Location','Electric Utility'],axis =1)
```
