#### REMOVING OUTLIERS

```python
Q1 = df.Energy.quantile(0.25)
print(Q1)
Q3 = df.Energy.quantile(0.75)
print(Q3)
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
lower_limit,upper_limit
```

```python
df[(df.Energy < lower_limit) | (df.Energy > upper_limit)]
```

```python
df=df[(df.Energy < lower_limit) | (df.Energy > upper_limit)]
df
```

```python
df.shape
```

```python
dummies = pd.get_dummies(df[["Time","BS"]], drop_first=True)
dummies
```

```python
df1 = pd.concat([df, dummies], axis=1)
df1.drop(columns=["Time","BS"], inplace=True)
```

```python
df1
```

#### TRAIN TEST

```python
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop("Energy",axis=1)
y = df1["Energy"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x_train,y_train)
```

```python
model.predict(x_test).round(2)
```
